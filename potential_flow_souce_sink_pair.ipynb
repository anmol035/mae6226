# Potential flow plots for  source,sink and pair

```python
import math
import numpy
from matplotlib import pyplot
```

```python
M=60
x_begin=-2 
x_end=2
y_begin=-1 
y_end=1
```

```python
xx = numpy.linspace(x_begin,x_end)
yy = numpy.linspace(y_begin,y_end)
```

```python
xx,yy = numpy.meshgrid(x,y)
```

```python
%matplotlib inline 
thick=10.0
h=(y_begin-y_end)/(x_begin-x_end)*thick
pyplot.figure(figsize=(thick,h))
pyplot.xlabel("X",fontsize=16)
pyplot.ylabel("Y",fontsize=16)
pyplot.xlim(x_begin,x_end)
pyplot.ylim(y_begin,y_end)
pyplot.scatter(xx,yy,s=5,color="blue",marker='o')
```

## potential flow plot for source

```python
strength=5
x_source=-1.0
y_source=0.0
q=((xx-x_source)**2)+((yy-y_source)**2)
z=numpy.sqrt(q)
u_source=(strength/(2*math.pi))*(numpy.log1p(z))
v_source=(strength/(2*math.pi))*(numpy.log1p(z))
```

```python
thick=10.0
h=(y_begin-y_end)/(x_begin-x_end)*thick
pyplot.figure(figsize=(thick,h))
pyplot.xlabel("X",fontsize=16)
pyplot.ylabel("Y",fontsize=16)
pyplot.xlim(x_begin,x_end)
pyplot.ylim(y_begin,y_end)
pyplot.streamplot(xx,yy,u_source,v_source,density=2
                ,linewidth=1,arrowsize=2,arrowstyle='->')
pyplot.scatter(x_source,y_source,color='green',s=70,marker='o')
```

## potential flow plot for sink

```python
strength=-5
x_sin=1.0
y_sin=0.0
q=(xx-x_sin)**2+(yy-y_sin)**2
z=numpy.sqrt(q)
u_sin=(strength/(2*math.pi))*(numpy.log1p(z))
v_sin=(strength/(2*math.pi))*(numpy.log1p(z))
```

```python
thick=10.0
h=(y_begin-y_end)/(x_begin-x_end)*thick
pyplot.figure(figsize=(thick,h))
pyplot.xlabel("X",fontsize=16)
pyplot.ylabel("Y",fontsize=16)
pyplot.xlim(x_begin,x_end)
pyplot.ylim(y_begin,y_end)
pyplot.contour(xx,yy)

```

```python
u_final=u_source+u_sin
v_final=v_source+v_sin
thick=10.0
h=(y_begin-y_end)/(x_begin-x_end)*thick
pyplot.figure(figsize=(thick,h))
pyplot.xlabel("X",fontsize=16)
pyplot.ylabel("Y",fontsize=16)
pyplot.xlim(x_begin,x_end)
pyplot.ylim(y_begin,y_end)
pyplot.streamplot(x,y,u_final,v_final,density=3,linewidth=1,arrowsize=2,arrowstyle='->')
pyplot.scatter([x_source,x_sin],[y_source,y_sin],color='black',s=70,marker='o');
```
