# Potential flow plots for  source,sink and pair

```python
import math
import numpy
from matplotlib import pyplot
```

```python
M=60
x_begin=-2 
x_end=2
y_begin=-1 
y_end=1
```

```python
x = numpy.linspace(x_begin,x_end)
y = numpy.linspace(y_begin,y_end)
```

```python
xx,yy = numpy.meshgrid(x,y)
```

```python
%matplotlib inline 
thick=10.0
h=(y_begin-y_end)/(x_begin-x_end)*thick
pyplot.figure(figsize=(thick,h))
pyplot.xlabel("X",fontsize=16)
pyplot.ylabel("Y",fontsize=16)
pyplot.xlim(x_begin,x_end)
pyplot.ylim(y_begin,y_end)
pyplot.scatter(xx,yy,s=5,color="blue",marker='o')
```

## potential flow plot for source

```python
strength=5
x_source=-1.0
y_source=0.0
q=((xx-x_source)**2)+((yy-y_source)**2)
z=numpy.sqrt(q)
phi_source=(strength/(2*math.pi))*(numpy.log1p(z))
pyplot.contour(xx,yy,phi_source)
```

## potential flow plot for sink

```python
strength=-5
x_sin=1.0
y_sin=0.0

r=numpy.sqrt((xx-x_sin)**2+(yy-y_sin)**2)
phi_sink=(strength/(2*math.pi))*(numpy.log1p(r))
pyplot.contour(xx,yy,phi_sink)
```

```python

pyplot.contour(xx,yy,phi_sink+phi_source)
```
