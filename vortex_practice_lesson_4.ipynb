```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 50                                
x_start, x_end = -2.0, 2.0            
y_start, y_end = -1.0, 1.0            
x = numpy.linspace(x_start, x_end, N)    
y = numpy.linspace(y_start, y_end, N)    
X, Y = numpy.meshgrid(x, y)  
c=5.0
x_vor = 0.0 
y_vor = 0.0
```

```python
u = +c / (2 * math.pi) * (Y - y_vor) / ((X - x_vor)**2 + (Y - y_vor)**2)
v = -c / (2 * math.pi) * (X - x_vor) / ((X - x_vor)**2 + (Y - y_vor)**2)
psi = c / (4 * math.pi) * numpy.log((X - x_vor)**2 + (Y - y_vor)**2)
```

```python
#Plotting Vortex

width = 10
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v,
                  density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vor, y_vor, color='#CD2305', s=80, marker='o');
```

# Sink + Vortex

```python
c=-5.0 #circulation for sink
u_sink = +c / (2 * math.pi) * (Y - y_vor) / ((X - x_vor)**2 + (Y - y_vor)**2)
v_sink = -c / (2 * math.pi) * (X - x_vor) / ((X - x_vor)**2 + (Y - y_vor)**2)
psi_sink = c / (4 * math.pi) * numpy.log((X - x_vor)**2 + (Y - y_vor)**2)

# Now calculation total U,V and PSI

U=u_sink+u
V=v_sink+v
PSI=psi_sink+psi
```

```python
width = 10
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, U, V,
                  density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vor, y_vor, color='#CD2305', s=80, marker='o');
```
